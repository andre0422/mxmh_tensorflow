```python
dataset = dataset.drop(columns=['timestamp', 'permissions', 'primarystreamingservice', 'favgenre'])
dataset.columns
```

```python
dataset.describe()
```

Replace 'yes' and 'no' values into '1' and '0' values

```python
dataset = dataset.applymap(lambda x: 1 if x == 'Yes' else (0 if x == 'No' else x))
```

```python
dataset.info()
```

```python
dataset.head()
```

Noticed when I initially looked at the data in excel that the frequency responses vary for each genre. Need to investigate how many unique responses there are and turn them into the appropriate numerical values before we start exploring

```python
cols = ['freq_classical', 'freq_country', 'freq_edm', 'freq_folk', 'freq_gospel',
'freq_hiphop', 'freq_jazz', 'freq_kpop', 'freq_latin', 'freq_lofi',
'freq_metal', 'freq_pop', 'freq_rnb', 'freq_rap', 'freq_rock', 'freq_vidgame']

for col in cols:
    print(f"Unique values in {col}: {dataset[col].unique()}")
```

```python
dataset.musiceffects.unique()
```

Looks like the responses are consistent for each frequency column. We will now change the responses to numerical categorical data types

```python
mapping_freq = {'Very frequently': 3, 'Sometimes': 2, 'Rarely': 1, 'Never': 0}

for col in cols:
    dataset[col] = dataset[col].map(mapping_freq)
```

```python
mapping_effects = {'Improve': 2, 'Worsen' : 1, 'No effect': 0}

dataset.musiceffects = dataset.musiceffects.map(mapping_effects)
```

```python
for col in cols:
    print(f"Unique vals in {col}: {dataset[col].unique()}")
```

```python
dataset.musiceffects.unique()
```

Rounding values before we start some investigating

```python
dataset = dataset.round()
```

Using kNN to replace missing values 

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5, weights='uniform', metric='nan_euclidean')

dataset_imputed = imputer.fit_transform(dataset)

dataset_imputed = pd.DataFrame(dataset_imputed, columns=dataset.columns)

print(dataset_imputed)
```
