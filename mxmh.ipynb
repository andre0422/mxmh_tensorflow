Looks like the responses are consistent for each frequency column. We will now change the responses to numerical categorical data types

```python
mapping_freq = {'Very frequently': 3, 'Sometimes': 2, 'Rarely': 1, 'Never': 0}

for col in cols:
    dataset[col] = dataset[col].map(mapping_freq)
```
